```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv("sdf_20.csv")
df.head()
```

```python
def set_cl(row):
    return 0 if row['cl'] == 'X0' else 1

df_ev = df[df.ev == True]
df_ev['c'] = df.apply(lambda row: set_cl(row), axis=1)
df_train = df_ev.iloc[:,0:6]
df_target = df_ev.iloc[:, 8]
X = df_train.values
y = df_target.values

```

```python
X.shape
```

```python
from sklearn.model_selection import cross_validate, RepeatedKFold
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)

rkfold = RepeatedKFold(n_splits=5, n_repeats=3)

clf = RandomForestClassifier(random_state=42, n_estimators=100)
param_grid = {'criterion' : ['gini', 'entropy'], 'max_features' : ['auto', 'sqrt', 'log2', None]}
gs = GridSearchCV(estimator=clf, param_grid=param_grid, cv=rkfold)
gs.fit(X_train, y_train)

bs = gs.best_estimator_
y_true, y_pred = y_test, bs.predict(X_test)
print(classification_report(y_true, y_pred))
```

```python
import pickle
with open("./rf.pkl", 'wb') as f_out:
    pickle.dump(gs.best_estimator_, f_out)
```

```python
with open ("./rf.pkl", 'rb') as f_in:
    rf = pickle.load(f_in)

y_true, y_pred = y_test, rf.predict(X_test)
print(classification_report(y_true, y_pred))
```
